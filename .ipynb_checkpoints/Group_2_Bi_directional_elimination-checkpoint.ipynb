# Using wrap method to refine feature selection

###  Read clean data

```python
import pandas as pd

# Specify the path to your CSV file
file_path = 'Group_2_clean_Data..csv'

# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)

df.head()
```

```python
# Assuming `df` is your DataFrame:
df_feature = df.iloc[:, 5:-18]  # Select all rows and columns from index 5 to the 18th-to-last
df_target = df['violentPerPop']  # Select the 'violentPerPop' column as the target variable


df_feature.head(5)
```

# experimenting with backward feature selection

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import numpy as np

# Step 1: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    df_feature, df_target, test_size=0.2, random_state=42
)

# Step 2: Initialize the Base Model (Random Forest)
rf = RandomForestRegressor(random_state=42)

# Step 3: Recursive Feature Elimination with R-squared Monitoring
best_r2 = -np.inf  # Initialize the best R-squared score
best_num_features = None  # Track the best number of features
best_features = None  # Track the mask of the best features
num_features = X_train.shape[1]  # Total number of features
performance_tracker = []  # Track R-squared scores for each step

# Loop through different numbers of features
for n_features in range(1, num_features + 1):
    rfe = RFE(estimator=rf, n_features_to_select=n_features)
    rfe.fit(X_train, y_train)
    
    # Subset the dataset to include only selected features
    selected_columns = X_train.columns[rfe.support_]
    X_train_rfe = X_train[selected_columns]
    X_test_rfe = X_test[selected_columns]
    
    # Train and evaluate the model with the selected features
    rf.fit(X_train_rfe, y_train)
    y_pred = rf.predict(X_test_rfe)
    r2 = r2_score(y_test, y_pred)
    performance_tracker.append((n_features, r2))
    
    # Check for improvement in R-squared
    if r2 > best_r2:
        best_r2 = r2
        best_num_features = n_features
        best_features = rfe.support_
    else:
        # Stop if no improvement in R-squared
        print(f"No improvement in R-squared with {n_features} features. Stopping early.")
        break

# Step 4: Use the best features
final_selected_columns = X_train.columns[best_features]
print(f"Best Number of Features: {best_num_features}")
print(f"Best R-squared Score: {best_r2:.4f}")
print(f"Selected Features: {list(final_selected_columns)}")

# Step 5: Train the Final Model with Best Features
X_train_final = X_train[final_selected_columns]
X_test_final = X_test[final_selected_columns]
rf.fit(X_train_final, y_train)

# Predict and Evaluate Final Model
y_pred_final = rf.predict(X_test_final)
final_r2 = r2_score(y_test, y_pred_final)

print(f"Final Model Performance:")
print(f"  - Best R-squared: {final_r2:.4f}")

# Optional: Visualize the Performance Progression
import matplotlib.pyplot as plt
performance = np.array(performance_tracker)
plt.plot(performance[:, 0], performance[:, 1], marker='o')
plt.xlabel('Number of Features')
plt.ylabel('R-squared Score')
plt.title('R-squared vs. Number of Features')
plt.show()
```

#  experimenting  model with forward wrap selection

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    df_feature, df_target, test_size=0.2, random_state=42
)

# Step 2: Initialize the Base Model (Random Forest)
rf = RandomForestRegressor(random_state=42)

# Step 3: Forward Feature Selection with Tracking
performance_tracker = []  # Track (number of features, R-squared) at each step
selected_features = []  # Track selected features

# Loop through features using forward selection
for num_features in range(1, X_train.shape[1] + 1):
    sfs = SequentialFeatureSelector(
        estimator=rf,
        n_features_to_select=num_features,  # Add one feature at a time
        direction="forward",  # Forward selection
        scoring="r2",  # Use R-squared for evaluation
        cv=5  # Cross-validation for robust performance
    )
    
    # Fit Sequential Feature Selector
    sfs.fit(X_train, y_train)
    
    # Get the selected feature mask
    selected_mask = sfs.get_support()
    selected_columns = X_train.columns[selected_mask]
    
    # Train and evaluate the model with selected features
    rf.fit(X_train[selected_columns], y_train)
    y_pred = rf.predict(X_test[selected_columns])
    r2 = r2_score(y_test, y_pred)
    
    # Track performance
    performance_tracker.append((num_features, r2))
    selected_features = list(selected_columns)
    
    # Stop if no significant improvement (optional)
    if len(performance_tracker) > 1 and r2 <= performance_tracker[-2][1]:
        print(f"No improvement in R-squared with {num_features} features. Stopping early.")
        break

# Step 4: Visualize the Performance Progression
performance = np.array(performance_tracker)
plt.plot(performance[:, 0], performance[:, 1], marker='o')
plt.xlabel('Number of Features')
plt.ylabel('R-squared Score')
plt.title('R-squared vs. Number of Features')
plt.grid()
plt.show()

# Step 5: Use the Best Features
best_num_features = int(performance[np.argmax(performance[:, 1]), 0])
best_r2 = performance[np.argmax(performance[:, 1]), 1]
final_selected_columns = selected_features[:best_num_features]

print(f"Best Number of Features: {best_num_features}")
print(f"Best R-squared Score: {best_r2:.4f}")
print(f"Selected Features: {final_selected_columns}")

# Step 6: Train Final Model with Best Features
X_train_final = X_train[final_selected_columns]
X_test_final = X_test[final_selected_columns]
rf.fit(X_train_final, y_train)

# Predict and Evaluate Final Model
y_pred_final = rf.predict(X_test_final)
final_r2 = r2_score(y_test, y_pred_final)
print(f"Final Model Performance:")
print(f"  - Best R-squared: {final_r2:.4f}")
```
